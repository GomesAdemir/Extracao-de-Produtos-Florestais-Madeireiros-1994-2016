```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lista = pd.read_csv('extracaopfm1994-2016.csv', delimiter=';',  on_bad_lines='skip')
```

```python
Tratamento = lista[['Tipo de produto florestal', 'Tipo de Floresta', 'Tipo de Uso',
                    'Unidade de Medida transformada', 'Ano_PEVS', 'Quantidade PEVS',
                    'Valor (mil reais)', 'Tipo de produto extrativo']]
                    
                       
df = Tratamento[Tratamento['Tipo de Uso'].isin(['Árvores abatidas', 
                                                'Borrachas', 
                                                'Fibras', 
                                                'Madeira para Combustível',
                                                'Madeira para Combustível - Subdivisão',
                                                'Madeira para Indústria'
                                                'Madeira para Indústria - subdivisão',
                                                'Madeira para Indústria - Subdivisão II'])]
```

----------------------------------------- Listas para cada Tipo de Uso ---------------------------------------------------- 

```python
ArvoresAbatidas = df.loc[(df['Tipo de Uso']=='Árvores abatidas')]
Borrachas = df.loc[(df['Tipo de Uso']=='Borrachas')]
Fibras = df.loc[(df['Tipo de Uso']=='Fibras')]

MadeiraCombustivel = df.loc[(df['Tipo de Uso']=='Madeira para Combustível')]
MadeiraCombustivelSub = df.loc[(df['Tipo de Uso']=='Madeira para Combustível - Subdivisão')]

MadeiraIndustria = df.loc[(df['Tipo de Uso']=='Madeira para Indústria')]
MadeiraIndustriaSub = df.loc[(df['Tipo de Uso']=='Madeira para Indústria - subdivisão')]
MadeiraIndustriaSubII = df.loc[(df['Tipo de Uso']=='Madeira para Indústria - Subdivisão II')]
```

```python
#----------------------Analise de Madeira Para Combunstível / Lenhas Silvicultura e Extrativismo--------------------------
#Valor de Mercado

lenhaSilv = MadeiraCombustivel.loc[(MadeiraCombustivel['Tipo de produto extrativo'] =='Lenha (silvicultura)')]
lenhaExt = MadeiraCombustivel.loc[(MadeiraCombustivel['Tipo de produto extrativo'] =='Lenha (extrativismo)')]

plt.figure(figsize=(20,10))

plt.title('Análise de mercado de madeira para combunstível - Lenha 1994/2016', fontsize=27, family = 'Times New Roman')
plt.ylabel("Valor de mercado (mil reais)", fontsize=18, family = 'Times New Roman')
plt.xlabel("Anos de PEV's", fontsize=18,  family = 'Times New Roman')
#Pontos-----------------
plt.scatter(lenhaSilv['Ano_PEVS'], lenhaSilv['Valor (mil reais)'], label ='Lenha de Silvicutura', color = 'green')
plt.scatter(lenhaExt['Ano_PEVS'], lenhaExt['Valor (mil reais)'], label = 'Lenha de Extrativismo', color = 'blue')
#-----------------------
plt.plot(lenhaSilv['Ano_PEVS'], lenhaSilv['Valor (mil reais)'], label ='Lenha de Silvicutura', color = 'green')
plt.plot(lenhaExt['Ano_PEVS'], lenhaExt['Valor (mil reais)'], label = 'Lenha de Extrativismo', color = 'blue')
plt.legend(fontsize=13, loc = 6)
plt.show()
```

```python
#----------------------Analise de Madeira Para Combunstível / Carvão Vegetal-------------------------

CarvaoExtrat = MadeiraCombustivel.loc[(MadeiraCombustivel['Tipo de produto extrativo'] =='Carvão vegetal (extrativismo)')]
CarvaoSilvi = MadeiraCombustivel.loc[(MadeiraCombustivel['Tipo de produto extrativo'] =='Carvão vegetal (silvicultura)')]

plt.figure(figsize=(20,10))
plt.title('Análise de mercado de madeira para combunstível - Carvão Vegetal 1994/2016', fontsize=27, family = 'Times New Roman')
plt.ylabel("Valor de mercado (mil reais)", fontsize=18, family = 'Times New Roman')
plt.xlabel("Anos de PEV's", fontsize=18,  family = 'Times New Roman')



plt.plot(CarvaoSilvi['Ano_PEVS'], CarvaoSilvi['Valor (mil reais)'], color ='green', label ='Carvão de Silvicutura')
plt.plot(CarvaoExtrat['Ano_PEVS'], CarvaoExtrat['Valor (mil reais)'], color ='blue', label = 'Carvão de Extrativismo')
plt.legend(fontsize=13, loc = 6)
plt.show()
```
